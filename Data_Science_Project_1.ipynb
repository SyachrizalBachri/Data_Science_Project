```python
# import library

import pandas as pd
import numpy as np
```

```python
# read file from csv file in gdrive 

df = pd.read_csv('/content/drive/MyDrive/Documents/File Data G_Colab/olist_orders_dataset.csv')
df.head()
```

```python
# data type for each column

df.info()
```

```python
# show columns in df

df.columns
```

```python
# change the data type of object to datetime 

dt_columns = ['order_purchase_timestamp','order_delivered_carrier_date','order_delivered_customer_date','order_estimated_delivery_date']

for i in dt_columns:
  df[i] = pd.to_datetime(df[i])
```

```python
# data type already change to datetime

df.info()
```

```python
# adding a year and month column from the order purchase timestamp column 

df['year'] = df['order_purchase_timestamp'].dt.year
df['month'] = df['order_purchase_timestamp'].dt.month_name()
```

```python
df.head()
```

```python
# count of order_id for each year and month

df.groupby(['year', 'month']).agg({'order_id' : 'count'}).unstack()
```

```python
# save into file_excel variabel

file_excel = df.groupby(['year', 'month']).agg({'order_id' : 'count'}).unstack()
```

```python
# save the pivot results as an excel file

file_excel.to_excel('pivot.xlsx')
```
